```python
Counter(dboth.get_label())
```

```python
params = {}
params['objective'] = "binary:logistic"
params['eta'] = 0.1 #0.05 #0.01
params['min_child_weight'] = 6
params['subsample'] = 0.7
params['colsample_bytree'] = 0.7
params['max_delta_step'] = 1
params['silent'] = 1
params['max_depth'] = 9 #7
params['eval_metric'] = 'auc'
# params["scale_pos_weight"] = 1.0
params['gamma'] = 5
params['alpha'] = 1
# params['nthread'] = 4

def fpreproc(dtrain, dtest, param):
    label = dtrain.get_label()
    ratio = float(np.sum(label == 0)) / np.sum(label==1)
    param['scale_pos_weight'] = ratio
    return (dtrain, dtest, param)

plst = list(params.items())

max_rounds = 7000
```

```python
# train_ind = np.load('../saved/stupid_encode/enc_ind.npy')
val_ind = np.load('../saved/stupid_encode/val_ind.npy')
```

```python
xfer_base_perc=0.05
rat_pos, rat_neg = [(dboth.get_label() == v).sum() / float(dboth.num_row()) for v in [1, 0]]

dval = dboth.slice(val_ind) # dont change this
y_val = dval.get_label()
watchlist = [
#     (dtrain.slice(enc_ind), 'train'),
    (dval, 'val'),
]

targets_all = dboth.get_label()
targets_all[val_ind] = -2
dboth.set_label(targets_all)
```

```python
for t in range(50):
    print t

    # Train with eval
    train_ind = np.where(targets_all > -1)[0]
    u_ind = np.where(targets_all == -1)[0]
    dboth.set_label(targets_all)
    
    print 'Training Distribution:', Counter(dboth.slice(train_ind).get_label())
    
    bst = xgb.train(params, dboth.slice(train_ind), max_rounds, 
                    evals=watchlist, 
                    early_stopping_rounds=50, 
                    verbose_eval=True)
    
    # Predict
    preds = bst.predict(dboth.slice(u_ind), ntree_limit=bst.best_iteration)

    # Positive candidates
    perc_top = np.percentile(preds, 100 - xfer_base_perc*rat_pos)
    xfer_pos_bools = preds > perc_top
    print '# Pos xfer:', sum(xfer_pos_bools), '@ thresh:', perc_top

    # Negative candidates
    perc_bot = np.percentile(preds, xfer_base_perc*rat_neg)
    xfer_neg_bools = preds < perc_bot
    print '# Neg xfer:', sum(xfer_neg_bools), '@ thresh:', perc_bot

    xfer_pos_inds = u_ind[xfer_pos_bools]
    xfer_neg_inds = u_ind[xfer_neg_bools]

    # Transfer
    targets_all[xfer_pos_inds] = 1
    targets_all[xfer_neg_inds] = 0
    
```
